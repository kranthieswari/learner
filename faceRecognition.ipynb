```
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
%matplotlib inline
```

```
image=cv2.imread('known/3-heros.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
face_locations=face_recognition.face_locations(image)
number_of_faces=len(face_locations)
print("Found {} face(s) in input image" .format(number_of_faces))
```

```
plt.imshow(image)
ax = plt.gca()
for face_location in face_locations:
  top, right, bottom, left = face_location
  x, y, w, h= left, top, right, bottom
  print("Face is located at pixel location Top: {}, Left:{}, Bottom:{}, Right:{}" .format(x,y,w,h))
  rect=Rectangle((x,y), w-x, h-y, fill=False, color='blue')
  ax.add_patch(rect)
plt.show()
```

```
image=cv2.imread('known/modi.jpg')
modi_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image=cv2.imread('known/putin.jpg')
putin_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image=cv2.imread('known/jinping.jpg')
jinping_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```
modi_encoding=face_recognition.face_encodings(modi_image)[0]
putin_encoding=face_recognition.face_encodings(putin_image)[0]
jinping_encoding=face_recognition.face_encodings(jinping_image)[0]
known_face_encodings=[
  modi_encoding,
  putin_encoding,
  jinping_encoding
]
```

```
image=cv2.imread('known/modi1.jpg')
unknown_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(unknown_img)
```

```
unknown_encodings=face_recognition.face_encodings(unknown_img)
```

```
from scipy.spatial import distance
```

```
for unknown_encoding in unknown_encodings:
  results=[]
  for known_face_encoding in known_face_encodings:
    d=distance.euclidean(known_face_encoding,unknown_encodings)
    print("Euclidean distance", d)
    results.append(d)
  threshold=0.6
  results=np.array(results) <= threshold
  name= "unknown"
  if results[0]:
    name = "Modi1"
  elif results[1]:
    name= " Putin"
  elif results[2]:
    name= "Jinping"
  print(f"Found {name} in the photo")

```

```
image=cv2.imread('known/amir.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
face_landmarks_list=face_recognition.face_landmarks(image)
  
```

```
import matplotlib.lines as mlines
from matplotlib.patches import Polygon
```

```
plt.imshow(image)
ax=plt.gca()
for face_landmarks in face_landmarks_list:
  left_eyebrow_pts=face_landmarks['left_eyebrow']
  pre_x,pre_y=left_eyebrow_pts[0]
  for (x,y) in left_eyebrow_pts[1:]:
    l=mlines.Line2D([pre_x,x],[pre_y,y],color='white')
    ax.add_line(l)
    pre_x,pre_y=x,y
  right_eyebrow_pts=face_landmarks['right_eyebrow']
  pre_x,pre_y=right_eyebrow_pts[0]
  for (x,y) in right_eyebrow_pts[1:]:
    l=mlines.Line2D([pre_x,x],[pre_y,y],color='black')
    ax.add_line(l)
    pre_x,pre_y=x,y
  P=Polygon(face_landmarks['top_lip'],facecolor='green', edgecolor='orangered')
  ax.add_patch(P)
  P=Polygon(face_landmarks['bottom_lip'],facecolor='green', edgecolor='orangered')
  ax.add_patch(P)
plt.show()
```
